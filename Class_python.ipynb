### `split()` 메서드
`split()` 메서드는 문자열을 특정 구분자(delimiter)를 기준으로 분리하여 문자열 리스트로 반환합니다.

- **사용법:** `string.split(separator, maxsplit)`
  - `separator` (선택 사항): 문자열을 분리할 기준이 되는 구분자입니다. 지정하지 않으면 기본적으로 공백(스페이스, 탭, 줄 바꿈 등)을 기준으로 분리합니다.
  - `maxsplit` (선택 사항): 분할할 최대 횟수입니다. 지정된 횟수만큼 분할한 후 나머지 문자열은 마지막 요소로 포함됩니다. 기본값은 -1로, 가능한 모든 분할을 수행합니다.

```python
text_comma = "사과,바나나,오렌지"
words_by_comma = text_comma.split(',')
print(f"쉼표로 분리: {words_by_comma}")

text_space = "파이썬은 재미있어"
words_by_space = text_space.split() # 공백을 기준으로 분리
print(f"공백으로 분리: {words_by_space}")

text_maxsplit = "하나 둘 셋 넷 다섯"
words_maxsplit = text_maxsplit.split(' ', 2) # 2번만 분리
print(f"maxsplit=2로 분리: {words_maxsplit}")
```

### `join()` 메서드
`join()` 메서드는 문자열 리스트(또는 다른 iterable)의 모든 요소를 특정 문자열을 구분자로 사용하여 하나의 문자열로 결합합니다.

- **사용법:** `separator.join(iterable)`
  - `separator`: 각 요소를 연결할 때 사용될 구분자 문자열입니다.
  - `iterable`: 결합할 문자열 요소를 포함하는 리스트, 튜플 등의 반복 가능한 객체입니다. `iterable`의 모든 요소는 문자열이어야 합니다.

```python
words = ['Hello', 'World', 'Python']

# 공백을 구분자로 사용하여 결합
joined_with_space = ' '.join(words)
print(f"공백으로 연결: {joined_with_space}")

# 하이픈을 구분자로 사용하여 결합
joined_with_dash = '-'.join(words)
print(f"하이픈으로 연결: {joined_with_dash}")

# 빈 문자열을 구분자로 사용하여 결합 (모든 요소를 붙여서)
joined_without_separator = ''.join(words)
print(f"구분자 없이 연결: {joined_without_separator}")
```
